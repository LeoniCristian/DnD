```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

class dungeon():
    rock = 2
    floor = 0
    wall = 1
    
    def __init__(self,board_size):
        self.board = np.zeros(np.add(board_size,[+1,+1]))

    """
    r = initial percentage of rock cells
    """
    def seed_map(self,r):
        while(self.board.sum()/(dungeon.rock*np.prod(self.board.shape))<r):
            pos = [np.random.randint(size = 1,low =0 ,high=self.board.shape[i]) for i in range(self.board.ndim)]
            self.board[pos[0],pos[1]] = dungeon.rock
    
    """
    n:  CA iterations.
    T:  neighborhood value threshold that defines a rock
    M:  Moore neighborhood size
    """
    def CA_PCG(self,n,T,M):
        for i in range(0,n):
            for r in range(1,self.board.shape[0]):
                for c in range(1,self.board.shape[1]):
                    n_val = self.board[r-1:r+2,c-1:c+2].sum()/dungeon.rock
                    if(n_val>=T):
                        self.board[r,c] = dungeon.rock
                    else:
                        self.board[r,c] = dungeon.floor

        self.board[0,:] = dungeon.wall
        self.board[:,0] = dungeon.wall                      
        self.board[-1,:] = dungeon.wall
        self.board[:,-1] = dungeon.wall                      
        
        for r in range(1,self.board.shape[0]):
            for c in range(1,self.board.shape[1]):
                if(self.board[r,c] == dungeon.rock and not np.all(self.board[r-1:r+2,c-1:c+2] )):
                    self.board[r,c] = dungeon.wall
        
        self.board = self.board[1:-1,1:-1]
        return self.board
    
    def show(self):
        plt.imshow(self.board)
        plt.show()
        
d = dungeon([50,50])
d.seed_map(.5)

game_map = d.CA_PCG(5,5,3)
#plt.figure(figsize=(10,10))
d.show()

game_map = pd.DataFrame(game_map)
```
